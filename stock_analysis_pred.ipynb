```python
import pandas as pd
import numpy as np
from decimal import Decimal
```

```python
df = pd.read_csv("download.csv")

df.shape
```

```python
for col in df.columns:
    df[col] = df[col].astype(str).str.replace(",", "", regex=False)

X = df[['Open', 'High', 'Low', 'Volume']]
y = df['Price']

X
```

#### Average Price: (High+Low)/2

#### Price Movement: To understand how a stock has moved over time, you might look at the difference between the open and close which is (Price Change=Close−Open), This helps you gauge whether the stock gained or lost value during the trading session.

#### Volatility: You can measure volatility using the high and low prices, Volatility=High−Low, A larger range indicates higher volatility, which might suggest greater risk.

#### Weighted Average Price: If you want a more nuanced average considering volume, you can calculate the volume-weighted average price (VWAP), VWAP= ∑(Price×Volume)/∑Volume, This helps in understanding the average price a stock traded at throughout the day, weighted by the volume of shares traded.
